```python
for i in range(0, len(null)): 
	if null[i] > 0 and null[i] < 10: 
		col_name = df.columns[i] 
		df[col_name].fillna(df[col_name].mean(), inplace=True) 

# Print dataframe 
print(df.to_string())
```

```python
def drop_duplicates(df, column):
    # Find the duplicate rows
    duplicate_rows = df[df.duplicated(column, keep=False)]

    # Find the number of unique values in the column
    num_unique_values = len(duplicate_rows[column].unique())

    # Find the number of duplicate rows
    num_duplicate_rows = len(duplicate_rows)

    # Calculate the percentage of rows that are duplicates
    pct_duplicate_rows = num_duplicate_rows / (num_duplicate_rows + num_unique_values)

    # If more than 50% of the rows are duplicates, drop the duplicate rows
    if pct_duplicate_rows > 0.5:
        return df.drop_duplicates(subset=column)
    else:
        return df
```
